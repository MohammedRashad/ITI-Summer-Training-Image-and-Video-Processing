```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
chess = cv2.imread("chess.jpeg")
cv2_imshow(chess)
```

```python
gray = cv2.cvtColor(chess, cv2.COLOR_BGR2GRAY) 
cv2_imshow(gray)
```

```python
image_flt = np.float32(gray) 

res =  cv2.cornerHarris(image_flt, 2, 3 , 0.04)

cv2_imshow(res)
```

```python
chess[res > 0.01 * res.max()] = [0,0,255]

cv2_imshow(chess)
```

# Blob Detection

```python
blobs = cv2.imread("/content/Screen Shot 2022-08-23 at 8.23.24 PM.png", 0 )

cv2_imshow(blobs)
```

```python
params = cv2.SimpleBlobDetector_Params()
```

```python
params.filterByArea = True
params.minArea = 300

params.filterByCircularity = True
params.minCircularity = 0.9

params.filterByConvexity = True
params.minConvexity = 0.2

params.filterByInertia = True
params.minInertiaRatio = 0.1

detector = cv2.SimpleBlobDetector_create(params)
keypoints = detector.detect(blobs)

blank = np.zeros((1, 1))

number_of_blobs = len(keypoints)
print("Blobs are = " , number_of_blobs)

text = "Blobs are = " + str(number_of_blobs)

final = cv2.drawKeypoints(blobs, keypoints, blank, (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.putText(final, text, (100,60), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,0,0), 2 )
cv2_imshow(final)

```
